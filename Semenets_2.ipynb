```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
```

```python
#data
center = [13.3,6,3.1,1.8,1.5]
b = [5.9,4,2,1.1,0.7]
bar = [20,10,5,3,2.2]
city = ["Нью Йорк","Берлін","Варшава","Бухарест","Київ"]
```

```python
fig, ax = plt.subplots(figsize = (8,10))
j = 0
for i in range(len(bar)):
  ax.bar(i,center[i],color = "sandybrown", label = city[i])
  ax.hlines(b[i],-0.5+j,0.5+j,color = "white",linewidth = 3.5)
  ax.text(i,center[i]-0.3,center[i],ha = "center",va = "center",color = "black",fontsize = 12)
  ax.text(i,b[i]-0.3,b[i],ha = "center",va = "center",color = "black",fontsize = 12)
  j+=1

ax.set_xticks(range(5))
ax.set_xticklabels(city)

ax.text(1,12,"Ціна за метр квадратний,",  fontsize=25)
ax.text(1,11.5,"тисяч доларів США",fontsize = 20)
ax.text(1, 11, "Дані: numbeo.com/cost-of-living", fontsize=13)


arrow1 = FancyArrowPatch(
    posA=(-0.2, 13), posB=(-0.8, 12.3),
    arrowstyle='->', mutation_scale=10, linewidth=0.7,
    connectionstyle="arc3,rad=0.35"
)
ax.add_patch(arrow1)
ax.text(-1.5, 12, "центр міста", fontsize=12, va='center')


arrow2 = FancyArrowPatch(
    posA=(-0.2, 5.6), posB=(-0.8, 4.7),
    arrowstyle='->', mutation_scale=10, linewidth=0.7,
    connectionstyle="arc3,rad=0.4"
)
ax.add_patch(arrow2)
ax.text(-1.7, 4, "окраїна міста", fontsize=12, va='center')









for a in ax.spines.values():
  a.set_visible(False)

ax.tick_params(axis = "both",length = 0)
ax.tick_params(axis = "y",labelleft = False)
ax.set_xlim(-1,5)
```
