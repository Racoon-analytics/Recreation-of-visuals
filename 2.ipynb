```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd


fig, ax = plt.subplots(figsize=(12, 8))
df = pd.DataFrame({'cities':[' ', '', '', '', ''], 'price':[19.2, 10.0, 5.1, 2.9, 2.2]})
top = [13.3, 6.0, 3.1, 1.8, 1.5]
bot = [5.9, 4.0, 2.0, 1.1, 0.7]
ax.bar(df['cities'], top, color='orange')

ax.text(1.2, 13, "   ,", fontsize=25, fontname='Arial')
ax.text(1.2, 12, "  ", fontsize=23, fontname='Arial')
ax.text(1.2, 11, ": numbeo.com/cost-of-living", fontsize=17, color='gray', fontname='Arial')

for i in range(5):
    ax.hlines(bot[i], i-0.5, i+0.5, colors='white', linewidth=5)


for i in range(5):
    ax.text(i, bot[i]-0.5, bot[i], horizontalalignment='center', fontsize=15)
    ax.text(i, top[i]-0.5, top[i], horizontalalignment='center', fontsize=15)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.get_yaxis().set_ticks([])
ax.xaxis.set_tick_params(length=0)
ax.tick_params(axis='x', labelsize=18)


style = "Simple, tail_width=0.5, head_width=4, head_length=8"
kw = dict(arrowstyle=style, color="k")
ar1 = patches.FancyArrowPatch((-0.2, 13.05), (-0.7, 12),
                             connectionstyle='angle3,angleA=0,angleB=90', **kw)
ax.text(-0.75, 11.6, " ", horizontalalignment='center', fontsize=12, fontname='Arial', alpha=0.8)

ar2 = patches.FancyArrowPatch((-0.2, 5.65), (-0.7, 4.5),
                             connectionstyle='angle3,angleA=0,angleB=90', **kw)

ax.text(-0.75, 4.2, " ", horizontalalignment='center', fontsize=12, fontname='Arial', alpha=0.8)

ax.add_patch(ar1)
ax.add_patch(ar2)









```
